### 1.5 Internal consistency: do SC + ST + Others add up to Total?

Since `cumul_hh_jobcards_tot` is reported separately from the three caste sub-categories, we can verify whether `sc + sts + others == tot`. Large discrepancies would directly undermine any equity analysis built on these variables.

```python
JobCards['components_sum'] = (
    JobCards['cumul_hh_jobcards_sc'] +
    JobCards['cumul_hh_jobcards_sts'] +
    JobCards['cumul_hh_jobcards_others']
)
JobCards['tot_discrepancy'] = JobCards['cumul_hh_jobcards_tot'] - JobCards['components_sum']

n_inconsistent = (JobCards['tot_discrepancy'].abs() > 0).sum()
print(f"Rows where sc + sts + others != tot: {n_inconsistent} ({100*n_inconsistent/len(JobCards):.2f}%)")
print()
print("Distribution of discrepancy (tot - sum of components):")
print(JobCards['tot_discrepancy'].describe())
```

Mostly the discrepancies are small (median is 0, mean is around 2), but there are some large outliers (max discrepancy is around 400) which may warrant further investigation.

Let's sort out the rows where the discrepancies are greater than 5% of the total number reported

```python
JobCards['gap_pct'] = (
    (JobCards['tot_discrepancy'] / JobCards['cumul_hh_jobcards_tot']) * 100
)

inconsistent = (JobCards['gap_pct'].abs() > 0.05).sum()
print(f"Rows where inconsistency is greater than 5% : {inconsistent} ({100*inconsistent/len(JobCards):.2f}%)")
```

A very small % of PCs display discrepancies greater than 5%. Therefore, we can just drop these rows for our analysis.

```python
JobCards_filtered = JobCards[(JobCards['gap_pct'].abs() <= 0.05)].drop(
                                 columns = ['components_sum', 'tot_discrepancy', 'gap_pct']
                             )
JobCards_filtered.head()
```

### 1.6 Monotonicity of cumulative variables over time

Since the job card variables are cumulative counts, they should be non-decreasing for the same GP across years. A year-on-year decline signals an administrative reset, data entry error, or GP boundary change — all of which are important to flag before using these as outcome variables.

```python
gp_key = ['state_name', 'district_name', 'block_name', 'gp_name']
cumul_cols = ['cumul_hh_jobcards_sc', 'cumul_hh_jobcards_sts',
              'cumul_hh_jobcards_others', 'cumul_hh_jobcards_tot']

# Sort by GP then year, compute year-on-year differences within each GP
JobCards_sorted = JobCards_filtered.sort_values(gp_key + ['year'])
diffs = JobCards_sorted.groupby(gp_key)[cumul_cols].diff()

# A violation is any row where any cumulative column decreased
violations = (diffs < 0).any(axis=1)
n_violations = violations.sum()
print(f"GP-year rows with at least one cumulative column decreasing: {n_violations} ({100*n_violations/len(JobCards):.2f}%)")
print()

# Break down by which column is violated most
print("Violations per column:")
print((diffs < 0).sum())
```

```python
# Inspect a sample of violating rows alongside their previous year
violating_idx = JobCards_sorted[violations].index
sample_violations = JobCards_sorted.loc[violating_idx].head(10)[
    gp_key + ['year'] + cumul_cols
]
display(sample_violations)
```

### 2.3 Dealing with duplicates

Let's inspect the numeric columns for the duplicated rows to understand what type of duplication we're dealing with — the same three cases can arise here as in Dataset 1 (zero ghost rows, split reporting, or conflicting full values).

```python
worker_numcols = [
    'applied_jobcards', 'issued_jobcards',
    'registered_workers_scs', 'registered_workers_sts',
    'registered_workers_others', 'total_registered_workers',
    'registered_workers_women',
    'active_jobcards',
    'active_workers_scs', 'active_workers_sts',
    'active_workers_others', 'total_active_workers',
    'active_workers_women'
]

Workers_dup[['year','state_name','district_name','block_name','gp_name'] + worker_numcols].head(20)
```

Let's look at a few specific GPs to classify the duplication pattern, just as we did for Dataset 1.

```python
# Pull the first three duplicate GP keys to inspect their patterns
first_dup_keys = (
    Workers_dup[['state_name','district_name','block_name','gp_name','year']]
    .drop_duplicates(subset=['state_name','district_name','block_name','gp_name'])
    .head(3)
)

for _, row in first_dup_keys.iterrows():
    mask = (
        (Workers['state_name']   == row['state_name'])  &
        (Workers['district_name']== row['district_name']) &
        (Workers['block_name']   == row['block_name'])  &
        (Workers['gp_name']      == row['gp_name'])
    )
    display(Workers[mask][['year','state_name','district_name','block_name','gp_name'] + worker_numcols])
```

These results reveal the same three duplication patterns as Dataset 1.

**Case 1 (Mathow, Arunachal Pradesh):** A real row paired with an all-zero ghost row, consistently across every year. The zero row likely represents a residual administrative entry with no actual data. The zero-row filter removes it cleanly.

**Case 2 (No Panchayat, Assam – Karbi Anglong):** One large dominant row paired with a small but non-zero satellite row (e.g. 6 applied job cards vs 14,760). The satellite is not all-zero so the filter alone does not drop it. However, since the dominant row is strictly larger on every column, `max()` effectively preserves the main row in full — the satellite contributes nothing.

**Case 3 (Chitvel, Andhra Pradesh):** Two rows with substantially different and plausible caste breakdowns — neither is a zero ghost nor a clear satellite. This is genuine split reporting: two administrative sub-units recorded under the same GP name. Taking `max()` column-wise mixes values from different rows (e.g. it would pair the larger SC count from one row with the larger Others count from another), which risks creating internally inconsistent caste totals. This is a known limitation of the `max()` approach and should be kept in mind when interpreting caste-disaggregated outcomes for affected GPs.

Applying the same two-step resolution used for Dataset 1: first drop all-zero rows (ghost entries), then take `max()` within each GP-year group for any remaining duplicates. Since the worker counts accumulate over a reporting period, the maximum is the best approximation of the true final value.

```python
Workers_clean = Workers.copy()

# Step 1: drop rows where all numeric columns are zero
Workers_clean = Workers_clean[Workers_clean[worker_numcols].sum(axis=1) > 0]
print(f"Rows after dropping all-zero rows: {len(Workers_clean)}")

# Step 2: collapse remaining duplicates by taking max within each GP-year
gp_key = ['year','state_name','district_name','block_name','gp_name']
Workers_clean = Workers_clean.groupby(gp_key, as_index=False)[worker_numcols].max()
print(f"Rows after deduplication: {len(Workers_clean)}")
```

```python
# Confirm all keys are now unique
Workers_clean[['year','state_name','district_name','block_name','gp_name']].value_counts().unique()
```

Now all keys are unique!

### 3.3 Data Quality Issue: Corrupted Rows in Mapping File

`PC_MIS_mapping.info()` shows all columns read as `object` type, which is unexpected for the ID and census code columns. Inspection reveals **420 rows where every column contains a GPS coordinate pair** (e.g. `87.20157 25.82769`) instead of the expected census identifiers. These rows appear to be GPS traces that were incorrectly appended to the CSV during file creation — every field in such a row is a `longitude latitude` string rather than a panchayat ID or name.

**Fix:** Drop all rows where `Panchayat_ID` cannot be parsed as a number, then cast `Panchayat_ID` and the three census code columns to `Int64`.

```python
# Drop corrupted GPS-coordinate rows
PC_MIS_mapping = PC_MIS_mapping[
    pd.to_numeric(PC_MIS_mapping['Panchayat_ID'], errors='coerce').notna()
].copy()

# Cast ID and census code columns to integer
int_cols = ['Panchayat_ID', 'Subdistrict census code', 'District census code', 'State census code']
PC_MIS_mapping[int_cols] = PC_MIS_mapping[int_cols].apply(pd.to_numeric).astype('Int64')

print(f'Rows after dropping corrupted entries: {len(PC_MIS_mapping):,}')
PC_MIS_mapping.info()
```

#### Issue 1: Exact Duplicate Rows

```python
#dplicated PC ids
dup_ids = PC_MIS_mapping[PC_MIS_mapping.duplicated(subset=['Panchayat_ID'], keep=False)]

print(f"Duplicates: {dup_ids.shape[0]}")          # total duplicate rows (should be ~16356)
print(f"Unique IDs: {dup_ids['Panchayat_ID'].nunique()}")   # unique IDs (~8178)
```

Seems like the duplicates might be perfect copies. Let's verify this.

```python
dup_ids.sort_values('Panchayat_ID').head(10)
```

16,356 rows share a duplicated `Panchayat_ID` (8,178 affected IDs). Every duplicate pair is **identical across all columns**, including `mis_index` — these are not conflicting records but verbatim copies, likely from double-appending a data block during file creation. They are safe to drop with `drop_duplicates()`.

```python
PC_MIS_mapping = PC_MIS_mapping.drop_duplicates(subset='Panchayat_ID').copy()
print(f'Rows after dropping exact duplicates: {len(PC_MIS_mapping):,}')
print(f'Unique Panchayat_IDs: {PC_MIS_mapping["Panchayat_ID"].nunique():,}')
```

#### Issue 2: `mis_index` Still Stored as Object

After the GPS-row removal, `mis_index` remains `object` type because the column contained mixed strings and numbers before cleaning. It should be cast to `float`. The ~47% missingness is expected — only 115k of 217k panchayats have a non-null MIS index (documented in CLAUDE.md).

```python
PC_MIS_mapping['mis_index'] = pd.to_numeric(PC_MIS_mapping['mis_index'], errors='coerce')
print(f'mis_index dtype: {PC_MIS_mapping["mis_index"].dtype}')
print(f'Missing mis_index: {PC_MIS_mapping["mis_index"].isna().sum():,} ({100*PC_MIS_mapping["mis_index"].isna().mean():.1f}%)')
```

#### Issue 3: Inconsistent String Formatting

`Gram_panchayat_name` has 101 entries with leading or trailing whitespace, and `Subdistrict name` is in mixed/title case while `District name` and `State name` are already lowercase. Stripping whitespace and lowercasing all name columns ensures consistent matching when this file is joined to other datasets.

```python
str_cols = ['Gram_panchayat_name', 'Subdistrict name', 'District name', 'State name']
PC_MIS_mapping[str_cols] = (
    PC_MIS_mapping[str_cols].apply(lambda s: s.str.strip().str.lower())
)
PC_MIS_mapping.info()
```

### 4.2 Mapping census data to state, district, and subdistrict names

**Approach:** `shrid_subdistrict` maps each village (`shrid2`) to internal SHRUG numeric IDs (`pc01_state_id`, `pc01_district_id`, `pc01_subdistrict_id`). Note these are **not** the same as the census codes in `PC_MIS_mapping` — shrid uses district-relative subdistrict numbering (1, 2, 3 per district) while PC_MIS uses state-relative codes (e.g. 200, 201...). The common key for the eventual merge into `panchayat_inequity` will therefore be **string names** (state, district, subdistrict), not numeric IDs.

`shrid_loc` maps the same `shrid2` to human-readable names. Joining these two on `shrid2` and collapsing to the subdistrict level gives a name-keyed lookup. That lookup is then merged with `pop_census_2001` (keyed by the same SHRUG numeric IDs) to attach population data to named subdistricts.

**Name conflicts:** 919 subdistricts have more than one `subdistrict_name` across their villages in `shrid_loc` (minor spelling variants). The most frequent name per subdistrict is taken as canonical. 6,701 villages in `shrid_subdistrict` have no match in `shrid_loc`; these contribute NaN names and are handled with a fallback in the aggregation.

```python
subdist_key = ['pc01_state_id', 'pc01_district_id', 'pc01_subdistrict_id']

# Step 1: join village-level shrids to their names
shrid_merged = shrid_subdistrict.merge(
    shrid_loc[['shrid2', 'state_name', 'district_name', 'subdistrict_name']],
    on='shrid2', how='left'
)

# Step 2: collapse to subdistrict level
# mode() raises IndexError when all values in a group are NaN, so we use a safe wrapper
def safe_mode(x):
    m = x.dropna().mode()
    return m.iloc[0] if len(m) > 0 else None

subdist_names = (
    shrid_merged
    .groupby(subdist_key)[['state_name', 'district_name', 'subdistrict_name']]
    .agg(safe_mode)
    .reset_index()
)

print(f'Unique subdistricts in name lookup: {len(subdist_names):,}')
subdist_names.head()
```

Merge the subdistrict name lookup with `pop_census_2001` on the three SHRUG numeric ID columns. All 5,333 subdistricts in `pop_census_2001` are present in `shrid_subdistrict`, so we expect a complete match. The result `subdist_census_2001` is keyed by names and ready to merge into `panchayat_inequity` via `PC_MIS_mapping` on `(state_name, district_name, subdistrict_name)`.

```python
# Step 3: attach census data to named subdistricts
subdist_census_2001 = subdist_names.merge(pop_census_2001, on=subdist_key, how='right')

unmatched = subdist_census_2001['subdistrict_name'].isna().sum()
print(f'Rows in subdist_census_2001: {len(subdist_census_2001):,}')
print(f'Unmatched (no name found): {unmatched}')
subdist_census_2001.head()
```

## 5. Merging Census Data with Inequity index on subdistricts

First let's select only relevant columns from the census data.

### 5.1 Merging `panchayat_inequity` with `subdist_census_2001_filtered`

The merge chain is:
1. Join `panchayat_inequity` → `PC_MIS_mapping` on `mis_index` to recover subdistrict names for each panchayat
2. Join that result → `subdist_census_2001_filtered` on `(state_name, district_name, subdistrict_name)`

**Pre-merge fix:** 31 name triplets appear more than once in `subdist_census_2001_filtered` because multiple census subdistrict IDs share the same name within a district (mostly urban Delhi subdivisions). These are aggregated by summing population columns before the merge to avoid row fan-out.

**Known data gap:** ~16,833 panchayats in `panchayat_inequity` have no match in `PC_MIS_mapping` (the mapping file only covers ~53% of panchayats), so those rows will have NaN census values.

```python
name_key = ['state_name', 'district_name', 'subdistrict_name']

# Collapse duplicate name triplets in census data by summing populations
subdist_census_agg = (
    subdist_census_2001_filtered
    .groupby(name_key, dropna=False)[['tot_pop_persons', 'SC_pop_persons', 'ST_pop_persons']]
    .sum()
    .reset_index()
)

# Step 1: attach subdistrict names to panchayat_inequity via mis_index
mis_to_loc = (
    PC_MIS_mapping[['mis_index', 'State name', 'District name', 'Subdistrict name']]
    .dropna(subset=['mis_index'])
    .drop_duplicates(subset='mis_index')
    .rename(columns={
        'State name': 'state_name',
        'District name': 'district_name',
        'Subdistrict name': 'subdistrict_name'
    })
)

panchayat_with_loc = panchayat_inequity.merge(mis_to_loc, on='mis_index', how='left')

panchayat_with_loc.head()
```

```python
# Step 2: attach census data on name triplet
panchayat_census = panchayat_with_loc.merge(
    subdist_census_agg, on=name_key, how='left', indicator=True
)

match_counts = panchayat_census['_merge'].value_counts()
print(f'Matched:    {match_counts.get("both", 0):,}')
print(f'Unmatched:  {match_counts.get("left_only", 0):,}')
print(f'Total rows: {len(panchayat_census):,}')

panchayat_census = panchayat_census.drop(columns='_merge')
panchayat_census.head()
```
